#### Одномерные данные

```python
fig = plt.figure(figsize= (14,5))
df['ChestPainType'].value_counts().plot.bar()
#Bar Chart (столбчатая диаграмма) - классический интсрумент для отображения информации по категориям
plt.show()
```

```python
fig = plt.figure(figsize= (14,5))
df['ChestPainType'].value_counts().sort_index().plot.bar()
plt.show()
```

```python
#Можно посмотреть относительные наблюдения
fig = plt.figure(figsize= (14,5))
(df['ChestPainType'].value_counts() / len(df)).plot.bar()
plt.show()
```

```python
df.groupby(['HeartDisease'])[['Age', 'Cholesterol']].mean()
```

```python
df.groupby(['HeartDisease'])[['Age','Cholesterol']].mean().plot.bar(figsize=(14,5),stacked = False) # stacked = False - два графика представляют отдельные столбцы
plt.show()
```

```python
df['Cholesterol_bin'] = df['Cholesterol']//10 * 10 #бинаризуем - разделим на определеннные бины, будто мы округлили каждое значение до 10
```

```python
#Используется при большом количестве вариантов категориальной переменной (более 20 значений) 
fig = plt.figure(figsize= (14,5))
chart = df['Cholesterol_bin'].value_counts().sort_index().plot.line()
plt.show()
```

```python
fig = plt.figure(figsize= (14,5))
chart = df['Cholesterol_bin'].value_counts().sort_index().plot.area(alpha = 0.5)
```

```python
fig = plt.figure(figsize= (14,5))
df['Age'].plot.hist(bins= 10)
#Гистограмма - классический инструмент для визуализации интервальных данных
#Визуально похожа на столбчатую диаграмму,гистограмма разбивает данные на равные интервалы и рисует частоты в каждом из столбцов
plt.show()
```

```python
fig = plt.figure(figsize= (14,5))
df['Age'].plot.hist(bins = 10) #с помощьюб параметра bins можно указать количество интервалов, по которым вы хотите провести визуализацию
```
